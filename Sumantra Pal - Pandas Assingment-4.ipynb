## Q3.- Write a function to return top-4 batsman at a venue. Take venue as input.

```python
def top_4_batsman(venue):
    new_data=match.merge(delivery,left_on='id',right_on='match_id')
    data=new_data[new_data['venue']==venue]
    return data.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).index[0:4]
```

```python
top_4_batsman('Rajiv Gandhi International Stadium, Uppal')
```
